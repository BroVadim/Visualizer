```python
import openpyxl
import json

class Visualizer:
    def __init__(self,file_path):
        self.file_path = file_path
    
    #добавить поле принадлежности адреса к первой или второй половине (для дальнейшей отрисовки)
    #НЕ НУЖНО НИЧЕГО ДОБАВЛЯТЬ, НУЖНО ПРОСТО ПРОВЕРИТЬ ПОЛЕ NEIGHBORS НА ПУСТОТУ
    #def searchObjects(self,num_address,num_square,num_latitude,num_longtitude):
    def searchObjects(self,num_address,num_latitude,num_longtitude,num_square):
        wb = openpyxl.load_workbook(self.file_path)
        sheet = wb['Лист1']
        addresses_list = []
        json = {}
        for count in range(2,sheet.max_row+1):
            cell = sheet.cell(row=count, column=num_address)
            address = str(cell.value)
            if address not in addresses_list:
                addresses_list.append(address)
                json[address] = {}
                
                cell = sheet.cell(row=count, column=num_latitude)
                json[address]["latitude"] = float(cell.value)
                
                cell = sheet.cell(row=count, column=num_longtitude)
                json[address]["longtitude"] = float(cell.value)
                
                cell = sheet.cell(row=count, column=num_square)
                json[address]["square"] = float(cell.value)
                
                json[address]["neighbors"] = []
                
                
            else:
                continue
        return json
    
    def searchNeighbors(self,addressesJson):
        wb = openpyxl.load_workbook(self.file_path)
        sheet = wb['Лист1']
        for count in range(2,sheet.max_row+1):
            cell = sheet.cell(row=count, column=1)
            address = cell.value
            cell = sheet.cell(row=count,column=6)
            neighbor_address = cell.value
            addressesJson[address]["neighbors"].append(neighbor_address)
        return addressesJson
        
        
    def createJson(self):
        file = open("map.js","w",encoding="utf-8")
        json1 = self.searchObjects(1,4,5,3)
        json2 = self.searchObjects(6,7,8,10)
        json1.update(json2)
        resultJson = self.searchNeighbors(json1)
        file.write("objects="+str(resultJson))
        file.close()

def printColumns(col1,col2,col3,col4):
    wb = openpyxl.load_workbook("map.xlsx")
    sheet = wb['Лист1']
    for count in range(2,sheet.max_row+1):
        cell = sheet.cell(row=count, column=col1)
        address = str(cell.value)
        cell = sheet.cell(row=count, column=col2)
        latitude = str(cell.value)
        cell = sheet.cell(row=count, column=col3)
        longtitude = str(cell.value)
        cell = sheet.cell(row=count, column=col4)
        square = str(cell.value)
        print(address+":"+"latitude: "+str(latitude)+":longtitude: "+str(longtitude)+":square: "+str(square))

if __name__=="__main__":
    vis = Visualizer("map.xlsx")
    vis.createJson()
    #printColumns(6,7,8,10)
    #printColumns(1,3,4,5)
    #printColumns(1,3,5,3)
```
